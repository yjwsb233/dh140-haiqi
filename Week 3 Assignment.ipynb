### Use a for loop to find if there is any country with no population

```python
no_pop = 0
for year in range(1963,2013):
    for i in range(len(pop[str(year)])):
        if pop[str(year)][i] == 0:
            no_pop = no_pop + 1
            print(pop['COUNTRY'][i])
    if year == 2012:
        break
```

This for loop allows us to traverse each row in each column from 1963 to 2012 to see if there is any country without a recorded population. If there is a country with 0 population, then the function will print the name of that country. It will also keep a record of the total number of countries without a recorded population and save it inside the variable `no_pop`.

```python
no_pop
```

There is no country with 0 recorded population, which means that the dataset is safe to put into use without much cleaning.

### Create a more advanced map based on the 1963 global population

```python
pop_web_mercator = pop.to_crs(epsg=3857)
# use subplots to set up the canvas
fig, ax = plt.subplots(figsize=(30,20))

# create the map plot
pop.plot(ax = ax,
         figsize=(30,20),
         column='1963',
         cmap='YlOrBr',
         legend=True,
         scheme='NaturalBreaks',
         legend_kwds={'loc':'lower right','bbox_to_anchor':(1,0),'ncol':5})

# add a title
ax.set_title('World Population Distribution in 1963')

# get rid of the axis
ax.axis('off')
ctx.add_basemap(ax,source=ctx.providers.CartoDB.Voyager)
#how to make map more in ratio? and how to make the legend bigger?
```

### Create a side-by-side map to show how the global population changed between 1963 and 2012

```python
# create the 1x2 subplots
fig, axs = plt.subplots(2, 1, figsize=(30, 20))

# name each subplot
ax1, ax2 = axs

# regular count map on the left
pop.plot(column='1963', 
            cmap='YlOrBr', 
            k=5, 
            edgecolor='white', 
            linewidth=0., 
            alpha=1, 
            ax=ax1, # this assigns the map to the subplot,
            legend=True)

ax1.axis("off")
ax1.set_title("1963 Population")

# spatial lag map on the right
pop.plot(column='2012', 
            cmap='YlOrBr', 
            k=5, 
            edgecolor='white', 
            linewidth=0., 
            alpha=1, 
            ax=ax2, # this assigns the map to the subplot
            legend=True
           )

ax2.axis("off")
ax2.set_title("2012 Population")
ctx.add_basemap(ax1,source=ctx.providers.CartoDB.Voyager)
ctx.add_basemap(ax2,source=ctx.providers.CartoDB.Voyager)

```

The comparison shows that there is a siginicant increase in the population of India. However, because the two color scales correspond to different values, it is difficult to get a sense of how much the population of each country has increased over these years. 

### Visualize the change in population of each country betweeen 1963 and 2012

```python
change = pop['2012'] - pop['1963']
change = pd.DataFrame(change)
change
```

#### Add `change` as a new column in the population dataset

```python
pop['change'] = change
```

#### Verify that `change` has been added as a new column

```python
pop.sample()
```

#### Create a map to visualize the change in population

```python
pop_web_mercator = pop.to_crs(epsg=3857)
# use subplots to set up the canvas
fig, ax = plt.subplots(figsize=(30,20))

# create the map plot
pop.plot(ax = ax,
         figsize=(30,20),
         column='change',
         cmap='YlOrBr',
         legend=True,
         scheme='NaturalBreaks',
         legend_kwds={'loc':'lower right','bbox_to_anchor':(1,0),'ncol':3})

# add a title
ax.set_title('Change in Population of Each Country from 1963 to 2012')

# get rid of the axis
ax.axis('off')
ctx.add_basemap(ax,source=ctx.providers.CartoDB.Voyager)
#how to make map more in ratio? and how to make the legend bigger?
```

Judging from the legend, it seems that several countries experienced a decrease in overall population between 1963 and 2012. I use the command belwo to find out which countries thess are.

#### Find out which countries experienced a population decline

```python
pop['COUNTRY'].where(pop['change'] < 0)
```

The command shows that Saint Kitts and Nevis, Hungary, Bulgaria, Georgia and Latvia saw a population decline.
